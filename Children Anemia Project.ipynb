# Converting the datatypes into integer to train the model

```python
# Converting the data types from object to integer
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Age_in_5year_groups'] = encoder.fit_transform(data['Age_in_5year_groups'])
data['Type_of_place_of_residence'] = encoder.fit_transform(data['Type_of_place_of_residence'])
data['Highest_educational_level'] = encoder.fit_transform(data['Highest_educational_level'])
data['Wealth_index_combined'] = encoder.fit_transform(data['Wealth_index_combined'])
data['Anemia'] = encoder.fit_transform(data['Anemia'])
data['Have_mosquito_bed_net_for_sleeping(from_household_questionnaire)'] = encoder.fit_transform(data['Have_mosquito_bed_net_for_sleeping(from_household_questionnaire)'])
data['Smokes_cigarettes'] = encoder.fit_transform(data['Smokes_cigarettes'])
data['Current_marital_status'] = encoder.fit_transform(data['Current_marital_status'])
data['Currently_residing_with_husband/partner'] = encoder.fit_transform(data['Currently_residing_with_husband/partner'])
data['When_child_put_to_breast'] = encoder.fit_transform(data['When_child_put_to_breast'])
data['Had_fever_in_last_two_weeks'] = encoder.fit_transform(data['Had_fever_in_last_two_weeks'])
data['Anemia_levell'] = encoder.fit_transform(data['Anemia_levell'])
data['Taking_iron_pills_sprinkles_or_syrup'] = encoder.fit_transform(data['Taking_iron_pills_sprinkles_or_syrup'])
```

```python
# Checkimg if the conversion is successful
data.dtypes
```

Now we are all set with data type

```python
# Checking for the missing values
data.isna().sum()
```

```python
# Filling the missing values
data['Hemoglobin_level adjusted_for_altitude_and_smoking(g/dl-1_decimal)'] = data['Hemoglobin_level adjusted_for_altitude_and_smoking(g/dl-1_decimal)'].fillna(data['Hemoglobin_level adjusted_for_altitude_and_smoking(g/dl-1_decimal)'].mean())
data['Hemoglobin_level_adjusted_for_altitude(g/dl-1decimal)'] = data['Hemoglobin_level_adjusted_for_altitude(g/dl-1decimal)'].fillna(data['Hemoglobin_level_adjusted_for_altitude(g/dl-1decimal)'].mean())
```
